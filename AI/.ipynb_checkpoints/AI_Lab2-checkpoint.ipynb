### 4.1 Implementation with graph input using BFS

```python
def subset_sum_bfs(graph, target):
    # initializing queue with every node in graph
    queue = [i for i in graph]
    # empty lists
    visited = []
    result = []
    # iterate through the queue until its empty
    while queue:
        # set node as first element in queue and add it to visited
        node = queue.pop(0)
        visited.append(node)
        print(f'\nCurrent Node : {node}')
        # check the neighbours of the node
        for neighbour in graph[node]:
            # check if the neighbour node is visited
            if neighbour not in visited:
                print(f'Checking if {node} + {neighbour} equals {target}')
                # check if sum of neighbour and node is the target value
                if node + neighbour == target:
                    print(f'{node} + {neighbour} = {target}')
                    # if true then add it to result
                    result.append([node, neighbour])
                if neighbour == target and [neighbour] not in result:
                    result.append([neighbour])
                #queue.append(node)
    return result
```

```python
def subset_sum_bfs(graph, target):
    # initializing queue with every node in graph
    queue = [i for i in graph]
    # empty lists
    visited = []
    sum = 0
    result = []
    temp = []
    # iterate through the queue until its empty
    for i in range(len(queue)):
        temp = []
        sum = queue[i]
        temp.append(queue[i])
        for j in range(len(queue)):
            if i == j:
                break
            #print('***', sum)
            sum += queue[j]
            #print('**',queue[i], queue[j], sum)
            if sum <= target:
                temp.append(queue[j])
            else :
                break
        print(temp)
        if sum == target:
            result.append(temp)
    while queue:
        temp = []
        node = queue.pop(0)
        sum = node
        visited.append(node)
        temp.append(node)
        print(f'\nCurrent Node : {node}')
        for neighbour in graph[node]:
            if neighbour not in visited:
                if neighbour == target:
                    result.append([neighbour])
                    break
                sum = sum+neighbour
                #print(neighbour, sum)
                if sum <= target:
                    temp.append(neighbour)
                else:
                    break
                if (sum == target):
                    result.append(temp)
        
    return result
```

### 4.2 Implementation with graph input using DFS

```python
def dfs(node, graph, target, visited, result):
    # if node is visited then exit the function
    if node in visited:
        return
    # add node to visited    
    visited.append(node)

    # iterate through the neighbours of node
    for neighbour in graph[node]:
        # check if the neighbour is visited or not
        if neighbour not in visited:
            # check if sum or node and neighbour is target
            print(f'Checking if {node} + {neighbour} equals {target}')
            if node + neighbour == target:
                print(f'{node} + {neighbour} = {target}')
                result.append([node, neighbour])
            # call dfs function for the neighbour to search its next nodes (recursion)
            dfs(neighbour, graph, target, visited, result)

def subset_sum_dfs(graph, target):
    # declare empty lists
    visited = []
    result = []
    # iterate through each node in the graph
    for node in graph:
        # check if graph is visited or not 
        print(f'\nCurrent Node : {node}')
        if node not in visited:
            # call dfs function to search that node
            dfs(node, graph, target, visited, result)
    return result
```

```python
# initialize graph
graph = {
    1: [2,3],
    2: [5],
    3: [7,6],
    4: [5],
    5: [2,4],
    6: [3],
    7: [3]
}

# set target value
target = 9

# call function to get result and print it
result = subset_sum_dfs(graph, target)

print("\nResult : ", result)
```

### 4.3 Implementation with Set input

```python
set1 = [5,3,4,2,1,6]

target = 9

result = []

# iterating through the set
for i in range(len(set1)):
    print()
    for j in range(i+1, len(set1)):
        print(f'Checking if {set1[i]} + {set1[j]} equals {target}\t(Sum = {set1[i] + set1[j]})')
        # adding 2 elements to each other to check if its the target value without repetition
        if set1[i] + set1[j] == target and ([set1[i], set1[j]] and [set1[j], set1[i]] not in result):
            print(f'\n{set1[i]} + {set1[j]} = {target}\n')
            # add elements to result
            result.append([set1[i], set1[j]])

print("Result : ", result)
```
