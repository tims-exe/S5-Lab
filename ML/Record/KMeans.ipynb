```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../datasets/IRIS_DATASET.csv')

x = df.iloc[:,:2].values

model = KMeans(
    n_clusters = 3,
    random_state = 0,
)

model.fit(x)

print(model.cluster_centers_)

labels = model.predict(x)

plt.scatter(x[:,0], x[:,1], c = labels, edgecolor = 'k')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c = 'r', s = 150)
```
