```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from kneed import KneeLocator
```

```python
dataset = pd.read_csv('datasets/IRIS_DATASET.csv')

X = dataset.iloc[:,:2].values
```

```python
kmeans = KMeans(
    n_clusters=3, 
    random_state=0,
    init='k-means++',
    n_init="auto"
)

kmeans.fit(X)

print(f'Centroids : \n{kmeans.cluster_centers_}')

labels = kmeans.predict(X)

plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', marker='o', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=150, c='red', marker='o', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering (First 2 Features)')
plt.show()
```

```python
inertia_values = []
k_values = range(1, 11) 

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init="auto")
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)  # Sum of squared distances to the nearest cluster center


kneedle = KneeLocator(k_values , inertia_values, curve='convex', direction='decreasing')
optimal_k = kneedle.elbow

plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.vlines(optimal_k, plt.ylim()[0], plt.ylim()[1], linestyles="dashed", colors="red", label=f'Optimal k = {optimal_k}')
plt.show()
```

```python
print(f"The optimal number of clusters (k) is: {optimal_k}")

kmeans = KMeans(
    n_clusters=optimal_k, 
    random_state=0,
    init='k-means++',
    n_init="auto"
)

kmeans.fit(X)

print(f'Centroids : \n{kmeans.cluster_centers_}')

labels = kmeans.predict(X)

plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', marker='o', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=150, c='red', marker='o', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering (First 2 Features)')
plt.show()
```
